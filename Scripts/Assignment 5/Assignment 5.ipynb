## Joining the 2 Datasets
I chose to do an inner merge on both the player and team column to remove the players that did not play in the playoffs. Additionally, this helps us track the stats for most recent team a player has played for.

Example Situations:

1. Player A gets traded mid season from Team A to Team B = Only Player A's stats from Team B will be kept

2. Player A played in the regular season but did not make the playoffs = player A stats are removed

Doing this makes the data perfect for comparing regular and playoff stats without the unessecary data or NULL values for players that did not make the playoffs.


```python
merged_data = pd.merge(reg, po, on=['Player', 'Tm'], how='inner')
```
