# Cleaning Script
Small Cleaning script to prep the data prior to analysis. This file cleans the playoff dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import numpy as np
import os
```

```python
# The idea of changing encoding and delimeter were found through ChatGPT as importing the data using the previous method showed errors.
po = pd.read_csv("../RawData/Playoffs.csv", encoding="ISO-8859-1", delimiter= ";")
```

## Rounding the Numeric columns to 2 decimal places

```python
po = po.round(2)
```

```python
po.head()
```

# Removing Secondary positions
For example if a player played both the Center (C) position and the Power Forward (PF) position it would only take their "primary" position. In this case, the first position listed will be thier "Primary" position

```python
po["Pos"] = po["Pos"].str.split("-").str[0]
```

```python
folder_name = "CleanData"

os.makedirs(folder_name, exist_ok=True)

cleaned_file_path = os.path.join(folder_name, "Playoffs_Cleaned.csv")
po.to_csv(cleaned_file_path, index=False)

print(f"File saved at: {cleaned_file_path}")
```
